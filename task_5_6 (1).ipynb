```python
import time
import random

def random_numbers():
  shtraf = 3
  while True:
    a = random.randint(1,100)
    b = random.randint(1,100)

    print(f'Новый пример: {a} + {b}?')
    answer = int(input())

    if answer == a + b:
      print('Поздравляю! Ваш ответ правильный')
      break
    else:
      print(f'Неверно! Штрафная пауза: {shtraf} секунды')
      time.sleep(shtraf)
      shtraf+=1

random_numbers()
```
